```python
import numpy as np
import pandas as pd
```

```python
a = 0.3
b = 0.05
sigma = 0.03
T = 5
r = 0.08
```

```python
B = []
A = []
ZCB_prices = []
for i in range(5):
    Bi = (1-np.exp(-a*(i+1)))/(a)
    B.append(Bi)
    Ai = np.exp(((B[i]-(i+1))*(a*a*b-0.5*sigma*sigma))/(a*a)-((sigma*sigma*B[i]*B[i])/(4*a)))
    A.append(Ai)
    ZCBi = A[i]*np.exp(-B[i]*r)
    ZCB_prices.append(ZCBi)
    
    
print(B)
print(A)
print(ZCB_prices)
    
```

```python
swap_rate = (1-ZCB_prices[4])/((ZCB_prices[0]+ZCB_prices[1]+ZCB_prices[2]+ZCB_prices[3]+ZCB_prices[4]))
```

```python
print(swap_rate*100)
```
