```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
a = 0.3
b = 0.05
sigma = 0.03
T = 5
r = 0.08
K = 900
```

```python
# Calculating ZCB prices for maturity  of 1,2,3,4,5 years.
B = []
A = []
ZCB_prices = []
for i in range(5):
    Bi = (1-np.exp(-a*(i+1)))/(a)
    B.append(Bi)
    Ai = np.exp(((B[i]-(i+1))*(a*a*b-0.5*sigma*sigma))/(a*a)-((sigma*sigma*B[i]*B[i])/(4*a)))
    A.append(Ai)
    ZCBi = A[i]*np.exp(-B[i]*r)
    ZCB_prices.append(ZCBi)
    
    
print(B)
print(A)
print(ZCB_prices)
```

```python
# Calculating spot rates of 1,2,3,4 years
for j in range(5):
    z = (1000)/(ZCB_prices[j]*1000)
    w = z**(1/(j+1))
    spot_rates = w-1
    print(spot_rates)
```

```python
# Calculating current price of ZCB having maturity of 4yrs
B4 = (1-np.exp(-a*1))/(a)
A4 = np.exp((((B4-1))*(a*a*b-0.5*sigma*sigma))/(a*a)-((sigma*sigma*B4*B4)/(4*a)))
ZCB4 = A4*np.exp(-B4*0.0686532262940549)
print(ZCB4*1000)
```

```python
F0 = ZCB4*1000
print(F0)
```

```python
# Calculating option volatility
sigma_option = sigma*np.sqrt((1-np.exp(-2*a*4))/(2*a))*B4
print(sigma_option)
```

```python
Numerator = (np.log(F0/K))+(sigma_option*sigma_option*0.5*4)
Denominator = sigma_option*np.sqrt(4)
```

```python
d1 = Numerator/Denominator
d2 = d1-(sigma_option*np.sqrt(4))
```

```python
Nd1 = norm.cdf(d1, loc=0, scale=1)
Nd2 = norm.cdf(d2, loc=0, scale=1)
```

```python
option_price = (F0*Nd1-K*Nd2)*np.exp(-4*0.0686532262940549)
print(option_price)
```
